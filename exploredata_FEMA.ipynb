Useful Variables List:  
STATEFIPS: state fips code.  
STATE: state name
POPULATION: state population.   
SOVI_SCORE: social vulnerability score.  
HRCN_EVNTS: # of hurricane events.  
HRCN_AFREQ: annualized hurricane frequencies.  
HRCN_EXPB: hurricane exposure: building value.  
HRCN_EXPP: hurricane exposure: population.



```
vlist = ['STATEFIPS', 'COUNTYFIPS','STATE','SOVI_SCORE','HRCN_EVNTS','HRCN_AFREQ','HRCN_EXPB','HRCN_EXPP','HRCN_RISKS']
```

```
county[vlist]
```

```
county[vlist].isna().sum()/len(county)
```

```
df_state = county[vlist]
```

```
df_fema_agg = df_state.groupby('STATEFIPS').agg({'SOVI_SCORE':'mean','HRCN_EVNTS':'max','HRCN_AFREQ':'mean','HRCN_EXPB':'sum','HRCN_EXPP':'sum','HRCN_RISKS':'mean'}).reset_index()
df_fema_agg.head()
```

```
df_fema_county = pd.DataFrame()
with open(path+'datasets/Data_Frames/df_fema_county.pkl','rb') as infile:
  df_fema_county = pickle.load(infile)
```

```
renamed = df_fema_county.rename(columns={'STATEFIPS': 'State Code', 'STATE':'State', 'COUNTY':'County', 'COUNTYFIPS': 'County Code','SOVI_SCORE':'Social Vul. Score','HRCN_EVNTS':'Hrcn Events',
                               'HRCN_AFREQ':'Hrcn Annual Freq.','HRCN_EXPB':'Hrcn Exp.Building',
                               'HRCN_EXPP':'Hrcn Exp.Popul.','HRCN_RISKS':'Hrcn Risk Score'})

renamed[renamed["State Code"] == 48]
```

```
with open(path+'datasets/Data_Frames/df_fema_agg.pkl','wb') as outfile:
  pickle.dump(df_fema_agg, outfile)
```

```
df_features = pd.DataFrame()
with open(path+'datasets/Data_Frames/df_features.pkl','rb') as infile:
  df_features = pickle.load(infile)
df_features.head()
```

```
df_features.rename(columns={'sta01fips':'STATEFIPS'}, inplace=True)
```

```
df_merge = df_fema_agg.merge(df_features, on='STATEFIPS')
df_merge = df_merge.fillna(0)
df_merge.head()
```
